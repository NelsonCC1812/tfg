```python
df = pd.DataFrame(columns=data.columns)
tmp = []


for row in data.iloc[::-1].itertuples(index=False):

	tmp.append([*row._asdict().values()])

	if not pd.isna(row.Nombre) :
		
		tmp = pd.DataFrame(tmp[::-1])

		tmp = pd.DataFrame([tmp.apply(lambda col: SEPARATOR.join([str(elm) for elm in col.dropna().tolist()]))])
		#tmp = pd.DataFrame([[ (None  if len(col.tolist()[0]) == 0  else (col.tolist()[0][0] if len(col.tolist()[0]) == 1 else col.tolist()[0])) for _, col in tmp.items()]])
		df.loc[df.shape[0]] = tmp.iloc[0].tolist()

		tmp = []
	

df = df.iloc[::-1].reset_index(drop=True)
df.head()
```
