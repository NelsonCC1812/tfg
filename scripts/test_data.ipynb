```python
from collections import namedtuple

columns = namedtuple('Columns', ' '.join(columns.keys()))(**columns)
years = namedtuple('Years', re.sub(r'[\[\],\']','', str([*map(lambda y: f'y{y}', years)])))(*years)

YYYY = lambda txt, year : txt.replace('[YYYY]', str(year))
```
